```python
def is_win(df):
    winner = []
    for item,row in df.iterrows():
        if row.winner == row.batting_team:
            winner.append(1)
        else:
            winner.append(0)

    return winner
            
```

```python
delivery_df['winner']=is_win(delivery_df)
```

```python
import random

# Mapping of ICC teams to their home cities
icc_cities_dict = {
    'South Africa': 'Johannesburg',
    'Afghanistan': 'Kabul',
    'Australia': 'Melbourne',
    'Netherlands': 'Amsterdam',
    'Pakistan': 'Lahore',
    'Scotland': 'Edinburgh',
    'West Indies': 'Kingston',
    'India': 'Mumbai',
    'United States of America': 'Lauderhill',
    'England': 'London',
    'Namibia': 'Windhoek',
    'New Zealand': 'Auckland',
    'Ireland': 'Dublin',
    'Uganda': 'Kampala',
    'Nepal': 'Kathmandu',
    'Bangladesh': 'Dhaka',
    'Oman': 'Muscat'
}

# Replace NaN values in 'city' with a placeholder (0)
final_df['city'] = final_df['city'].fillna(0)

# Function to fill missing city values based on batting or bowling team
def fill_icc_city(x):
    if x.city == 0:  # If city is missing
        team = random.choice([x.batting_team, x.bowling_team])  # Randomly select one team
        return icc_cities_dict.get(team, "Unknown")  # Get city from dictionary or return "Unknown"
    else:
        return x.city  # Keep existing city if available

# Apply function to update the city column
final_df['city'] = final_df.apply(fill_icc_city, axis=1)
```

```python
import numpy as np
import pandas as pd
import pickle
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier

# ✅ Sample Data - Replace with Your Actual Dataset
# X_train = pd.read_csv("your_data.csv")  # Load actual training data
# y_train = pd.read_csv("your_target.csv")  # Load actual target values

# ✅ Fix NaN and Infinity Issues
X_train.replace([np.inf, -np.inf], np.nan, inplace=True)  # Convert infinities to NaN
X_train.ffill(inplace=True)  # Forward fill to fill missing values
  # Fill NaN values

y_train.dropna(inplace=True)  # Remove NaN values in the target variable

# ✅ Identify Numeric and Categorical Columns
numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['object']).columns.tolist()

# ✅ Convert categorical columns to string (prevents errors)
X_train[categorical_features] = X_train[categorical_features].astype(str)

# ✅ Pipelines for Numeric and Categorical Data
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Replace missing values with mean
    ('scaler', StandardScaler())  # Standardize numerical features
])

categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Fill missing categorical values
    ('encoder', OneHotEncoder(handle_unknown='ignore'))  # One-hot encoding
])

# ✅ Combine Preprocessing
preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# ✅ Define Model Pipeline
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1))
])

# ✅ Fit the Model
pipe.fit(X_train, y_train)

# ✅ Save the trained pipeline
with open("pipe.pkl", "wb") as f:
    pickle.dump(pipe, f)

print("✅ Model trained and saved successfully!")
```
